已知：

$$
\begin{bmatrix}
    u \\ v \\ 1
\end{bmatrix} = \frac{1}{^cZ}K{}^cT_w{}^wP = \frac{1}{^cZ}
\begin{bmatrix}
    k_x & 0 & u_b & 0 \\
    0 & k_y & v_b & 0\\
    0 & 0 & 1 & 0\\
\end{bmatrix}
\begin{bmatrix}
    R & p \\
    0 & 1 \\
\end{bmatrix}_{4\times4}
\begin{bmatrix}
    ^wX \\
    ^wY \\
    ^wZ \\
    1
\end{bmatrix}
$$

则有：

$$
K^{-1}{}\begin{bmatrix}
    u \\ v \\ 1
\end{bmatrix} = {}^cT_w{}^wP =
\begin{bmatrix}
    R & p \\
    0 & 1 \\
\end{bmatrix}_{4\times4}\frac{1}{{}^cZ}
\begin{bmatrix}
    ^wX \\
    ^wY \\
    ^wZ \\
    1
\end{bmatrix}
$$
